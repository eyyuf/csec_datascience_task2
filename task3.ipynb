## Part 2 – Titanic Analysis

### Step 1 – Exploration

```python
df = pd.read_csv('dataset-2.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Step 2 – Data Cleaning

```python
# Fill Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Embarked with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin
df.drop(columns=['Cabin'], inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

print('Missing values after cleaning:')
print(df.isnull().sum())
```

### Step 3 – Data Analysis (groupby)

```python
# Survival rate by gender
df.groupby('Sex')['Survived'].mean()
```

```python
# Survival rate by class
df.groupby('Pclass')['Survived'].mean()
```

```python
# Average age per class
df.groupby('Pclass')['Age'].mean()
```

```python
# Survival rate by age group
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,19,59,120], labels=['Child','Teen','Adult','Senior'])
df.groupby('AgeGroup', observed=True)['Survived'].mean()
```

### Step 4 – Filtering

```python
# Female passengers who survived
female_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print(f'Female survivors: {len(female_survived)}')
female_survived[['Name','Age','Pclass']].head()
```

```python
# Children who survived (age <= 12)
children_survived = df[(df['Age'] <= 12) & (df['Survived'] == 1)]
print(f'Children survivors: {len(children_survived)}')
children_survived[['Name','Age','Sex','Pclass']].head()
```

```python
# 1st class passengers who survived
class1_survived = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
print(f'1st class survivors: {len(class1_survived)}')
class1_survived[['Name','Age','Sex']].head()
```

### Step 5 – Insights

```python
gender_rate  = df.groupby('Sex')['Survived'].mean()
class_rate   = df.groupby('Pclass')['Survived'].mean()
age_rate     = df.groupby('AgeGroup', observed=True)['Survived'].mean()
combo_rate   = df.groupby(['Sex','Pclass'])['Survived'].mean()

print('1. Gender – more likely to survive:')
print(f'   Females: {gender_rate["female"]:.1%}  |  Males: {gender_rate["male"]:.1%}')
print('   → Females were far more likely to survive.\n')

print('2. Did class affect survival?')
print(f'   1st: {class_rate[1]:.1%}  |  2nd: {class_rate[2]:.1%}  |  3rd: {class_rate[3]:.1%}')
print('   → Yes. Higher class = higher survival rate.\n')

print('3. Were children prioritized?')
print(f'   Children: {age_rate["Child"]:.1%}  |  Adults: {age_rate["Adult"]:.1%}')
print('   → Yes. Children had a higher survival rate than adults.\n')

best = combo_rate.idxmax()
print('4. Best combination:')
print(f'   {best[0].capitalize()} in Class {best[1]} → {combo_rate.max():.1%} survival rate')
```
